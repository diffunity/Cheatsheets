```python
from sklearn.model_selection import StratifiedShuffleSplit, KFold, StratifiedKFold

X = titanic_data.drop(["Survived"], axis = 1).copy()
X = X[X.columns[X.dtypes==int]]
y = titanic_data["Survived"].copy()

# Vanilla Split
from sklearn.model_selection import train_test_split
trainX, testX, train_y, test_y = train_test_split(X,y,test_size = 0.333, random_state = 333)

# KFold (test_size = 1/n_splits)
splitter = KFold(n_splits = 8, random_state = 333)
splitter.split(X, y)

# Stratified K fold
splitter = StratifiedKFold(n_splits = 8, random_state = 333)
splitter.split(X, y)

# Stratified Shuffle Split
splitter = StratifiedShuffleSplit(n_splits = 8, test_size = 0.333, random_state = 333)
splitter.split(X, y)
```

```python
## Ensemble for regressors

X = boston_data.drop(["id","price","date"],axis = 1).reset_index(drop = True)
y = boston_data["price"].copy()

import sklearn as sk
import sklearn.ensemble as sken
from sklearn.ensemble import *
from sklearn.linear_model import *
from sklearn.calibration import *
from sklearn.dummy import *
from sklearn.gaussian_process import *
from sklearn.neighbors import *
from sklearn.tree import *
from sklearn.naive_bayes import *
from sklearn.svm import *
from sklearn.neural_network import *

import xgboost as xgb
import lightgbm as lgbm

trainX, testX, train_y, test_y = train_test_split(X,y,test_size = 0.333, random_state = 333)



models = {'model1': DummyRegressor(strategy = 'mean'),
          'model2': AdaBoostRegressor(loss = 'square'),
          'model3': BaggingRegressor(),
          'model4': ExtraTreesRegressor(),
          'model5': GradientBoostingRegressor(loss = 'huber'),
          'model6': RandomForestRegressor(), 
          'model7': LinearRegression(),
          'model8': PassiveAggressiveRegressor(),
          'model9': KNeighborsRegressor(n_neighbors = 3),
          'model10': MLPRegressor(), 
          'model11': DecisionTreeRegressor(),
          'model12': ExtraTreeRegressor(),
          'model13': xgb.XGBRegressor(),
          'model14': lgbm.LGBMRegressor()}


weight = {}

def average_blending(data, models):
    xtrain, ytrain, xtest = data
    blended_pred = 0
    for i in models.values():
        ttt = i
        ttt.fit(xtrain, ytrain)
        prediction = ttt.predict(xtest)
        blended_pred += (prediction / len(models))
    return blended_pred

avg_blended = average_blending(data = [trainX, train_y, testX], models = models)

# here weights can be given in proportion to the level of accuracy each models give for the training set
# and many other ways.
# in this example, arbitrary weights are set
weights = dict()
for i in models.keys():
    weights[i] = int(i.split("model")[-1]) / sum(range(1,len(models)+1))

def weighted_blending(data, models, weights):
    xtrain, ytrain, xtest = data
    blended_pred = 0
    for model,weight in weights.items():
        ttt = models[model]
        ttt.fit(xtrain, ytrain)
        prediction = ttt.predict(xtest)
        blended_pred += (prediction * weight)
    return blended_pred

weighted_blended = weighted_blending(data = [trainX, train_y, testX], models = models, weights = weights)
```

```python
def rmse(i1,i2):
    return np.sqrt(sum((i1-i2) ** (2)))

print(rmse(avg_blended, test_y))
print(rmse(weighted_blended, test_y))
```
